### Q.2. In which year highest number of the TV Shows & Movies were released ? Show with Bar Graph.

```python
import matplotlib.pyplot as plt
```

```python
# Convert 'Release_Date' to datetime, handling mixed formats and invalid dates
data['Release_Date'] = pd.to_datetime(data['Release_Date'], errors='coerce')
```

```python
# Extract the year from the 'Release_Date' column
data['Release_Year'] = data['Release_Date'].dt.year
```

```python
release_counts = data['Release_Year'].value_counts()
```

```python
# Find the year with the highest releases
highest_release_year = release_counts.idxmax()
highest_releases = release_counts.max()
```

```python
print(f"Year with the highest releases: {highest_release_year} ({highest_releases} releases)")
```

```python
# Plot the bar graph
plt.figure(figsize=(10, 6))
release_counts.sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of TV Shows and Movies Released per Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Releases', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
