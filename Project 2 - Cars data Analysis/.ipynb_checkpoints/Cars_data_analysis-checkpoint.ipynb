# 1) Instruction ( For Data Cleaning ) 
- Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column. 

```python
cars.isnull().sum()
```

```python
cars['Length'].fillna(cars['Length'].mean()) .head(50)

cars['Cylinders'].fillna(cars['Cylinders'].mean(), inplace=True) 
```

```python
cars
```

# 3) Instruction ( Filtering )
- Show all the records where Origin is Asia or Europe.

```python
cars.head(3)
```

```python
cars[cars['Origin'].isin(['Asia' , 'Europe'])]
```

# 4) Instruction ( Removing unwanted records )
- Remove all the records (rows) where Weight is above 4000.

```python
cars.head(2)
```

```python
#In pandas, the ~ (tilde) operator is a bitwise NOT operator. When used with a pandas DataFrame or Series, it inverts a boolean condition, allowing you to negate it.

cars[ ~ (cars['Weight'] > 4000)]  #The ~ negates this condition, giving True for rows where Weight is less than or equal to 4000.
```

# 5) Instruction ( Applying function on a column )
- Increase all the values of 'MPG_City' column by 3.

```python
cars['MPG_new_value'] = cars['MPG_City'] + 3
```
